```python

import spacy
import pandas as pd

nlp=spacy.load("en_core_web_sm")

data=pd.read_excel("Civic_voice_dataset.xlsx")

data.head(5)

def preprocess_data(text):

    doc = nlp(str(text).lower())

    cleaned_text=[]
    for token in doc:

        if not token.is_stop and token.is_alpha:
            cleaned_text.append(token.text)
    return cleaned_text

preprocess_sentences=[preprocess_data(row) for row in data['QUERIES'] ]

preprocess_sentences[:5]


from gensim.models import Word2Vec

model=Word2Vec(
    sentences=preprocess_sentences,
    vector_size=10,
    window=2,
    min_count=1,
    sg=1,
    workers=4,
    epochs=10
)

import numpy as np

def sentence_vector(sentence):
    words = preprocess_data(sentence)

    vectors = []

    for word in words:
        if word in model.wv:
            vectors.append(model.wv[word])

    if len(vectors) == 0:
        return np.zeros(model.vector_size)

    return np.mean(vectors, axis=0)

X = np.array([sentence_vector(text) for text in data['QUERIES']])
y = data['URGENCY']
```

```python
import pandas as pd
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
data = {
    "complaint": [

        # WATER
        "water pipe burst and flooding",
        "no water supply in area",
        "water leakage from pipeline",
        "dirty water coming from tap",
        "low water pressure in houses",

        # ELECTRICITY
        "street light not working",
        "electric pole spark",
        "power outage in street",
        "electric wire hanging low",
        "transformer making loud noise",

        # ROAD
        "pothole on main road",
        "road completely damaged",
        "speed breaker broken",
        "road full of mud and water",
        "dangerous open manhole on road",

        # WASTE
        "garbage not collected",
        "overflowing garbage bin",
        "dead animal not removed",
        "waste dumped on roadside",
        "bad smell from garbage area",

        # OTHERS
        "stray dogs causing disturbance",
        "illegal construction blocking pathway",
        "public park damaged equipment",
        "noise pollution from loudspeakers",
        "public toilet not maintained"
    ],

    "urgency": [

        # WATER
        "High",
        "High",
        "High",
        "Medium",
        "Medium",

        # ELECTRICITY
        "Medium",
        "High",
        "High",
        "High",
        "Medium",

        # ROAD
        "High",
        "High",
        "Medium",
        "Medium",
        "High",

        # WASTE
        "Medium",
        "Medium",
        "High",
        "Medium",
        "Normal",

        # OTHERS
        "Medium",
        "High",
        "Medium",
        "Normal",
        "Medium"
    ]
}
df = pd.DataFrame(data)

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["complaint"])

y = df["urgency"]

model = LogisticRegression()
model.fit(X,y)

pickle.dump(model,open("urgency_model.pkl","wb"))
pickle.dump(vectorizer,open("urgency_vectorizer.pkl","wb"))

print("Urgency model saved")
```
